```python
import cv2
import numpy as np
```

```python
def grid(cnv, x_step, y_step, color=(0,0,0)):
    cnv[::y_step,:] = color
    cnv[:,::x_step] = color
```

```python
def draw(even, x,y,flags,param):
    x_=x//20
    y_=y//20
    x=x-x%20+1
    y=y-y%20+1
    try:
        if flags == 1:
            cnv[y:y+19,x:x+19] = (0,0,0)
            X_in [y_,x_] = 255
        if flags == 2:
            cnv[y:y+19,x:x+19] = (255,255,255)
            X_in [y_,x_] = 0
    except:
        pass
           
```

```python
X_in = np.zeros( (28,28), dtype=np.uint8() )
cnv = np.ones((560,560,3), dtype=np.uint8() ) *255
grid(cnv,20, 20,(100,200,0))

cv2.namedWindow("canvas")
cv2.setMouseCallback('canvas', draw)
while True:
    cv2.imshow("canvas", cnv)
    if cv2.waitKey(1) == 27:
        break
cv2.destroyAllWindows()
X_in
```
